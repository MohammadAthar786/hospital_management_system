






# 🏥 Healthcare Management System with ML, Algorithms & Cloud Deployment

## 🚀 Project Overview

This project is a **scalable Healthcare Management System backend** built using modern backend architecture, enhanced with:

* **Machine Learning models**
* **Algorithm-driven decision systems**
* **Secure authentication & authorization (JWT + RBAC)**
* **Cloud deployment on AWS**

It manages:

* Patient records
* Doctor workflows
* Appointments
* Resource allocation

while ensuring **security, scalability, and intelligent decision-making**.

---

## 🎯 Objectives

* Build a **production-ready backend system** using FastAPI
* Design **modular architecture** (Router → Service → Repository → Model)
* Integrate **Machine Learning APIs**
* Implement **real-world algorithms**
* Secure APIs using **JWT Authentication & RBAC**
* Deploy database on **Amazon Web Services RDS (PostgreSQL)**
* Containerize using Docker
* Deploy APIs on EC2

---

## 🧱 System Architecture

```
Client (Postman / Frontend / Streamlit)
            ↓
   FastAPI (API Layer + Auth Middleware)
            ↓
   Service Layer (Business Logic + Algorithms)
            ↓
   Repository Layer (DB Operations)
            ↓
   PostgreSQL (AWS RDS)

            ↓
   ML Models (Prediction APIs)

            ↓
Docker Container → AWS EC2 Deployment
```

## 🔐 Authentication & Authorization (NEW)

### 🔑 JWT Authentication

* Secure login system using **JSON Web Tokens**
* Token-based authentication for all protected APIs
* Stateless session handling (no server-side session storage)

**Flow:**

1. User logs in → receives JWT token
2. Token sent in request headers
3. Middleware validates token
4. Access granted/denied

---

### 👥 Role-Based Access Control (RBAC)

System defines roles such as:

* **Admin**
* **Doctor**
* **Patient**

**Access Control Examples:**

* Admin → Full access (users, resources, analytics)
* Doctor → Manage patients & appointments
* Patient → View own data & bookings

👉 Implemented via:

* Role checks in service layer
* Route-level authorization guards
